## Importing RF model and Cleaning of Data

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
place_dict = {
    "First" : 1,
    "Second" : 2,
    "Third" : 3,
    "Fourth" : 4
} 
```

```python
age_dict = {
    "20-29" : 2,
    "10-19" : 1,
    "30-39" : 3,
    "50-59" : 5,
    "60-69" : 1,
}
```

```python
df['Age'].value_counts()
```

```python
# converting 1st,2nd,3rd,4th to 1,2,3,4
for idx in range(len(df)):
    arr = df.iloc[idx,:]
    for idx2, val in enumerate(arr):
        if val in place_dict.keys():
            df.iloc[idx,idx2] = place_dict[val]
```

```python
# converting ages to ints
for idx in range(len(df)):
    arr = df.iloc[idx,:]
    for idx2, val in enumerate(arr):
        if val in age_dict.keys():
            df.iloc[idx,idx2] = age_dict[val]
```

```python
# filtering out time and email and comments (don't need personal information)
```

```python
filt_df = df[['Age', 
              'Med_Tempo', 'High_Tempo', 'Slow_Tempo', 
              'Monophonic', 'Polyphonic', 'Homophonic', 
              'Jazz_Monophonic', 'Jazz_Homophonic', 
              'Commuting', 'Working_Out', 'When_Cooking', 'Focus', 'Woodwinds', 'Brass', 'Percussions', 'Strings']]
```

```python
labels = np.array(filt_edited_commuting['Commuting_2'])  # Ensure labels is 1D
feature_list = ['Age', 'Med_Tempo', 'High_Tempo', 'Slow_Tempo', 'Monophonic',
       'Polyphonic', 'Homophonic', 'Jazz_Monophonic', 'Jazz_Homophonic', 'Woodwinds', 'Brass',
       'Percussions', 'Strings']  # List of feature columns

# Extract features as a 2D NumPy array
features = np.array(filt_edited_commuting[feature_list])

```

## RF Model + Feature Importance to Predict Where People Listen to Music

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.10, random_state=42
)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python

# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
feature_list
```

```python
# Extract feature importances
importances = rf.feature_importances_

# Create a DataFrame for better visualization
feature_names = train_features.columns if hasattr(train_features, 'columns') else [f"Feature_{i}" for i in range(train_features.shape[1])]
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

# Sort features by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Print the top features
print(feature_importances.head(5))
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
# errors = abs(predictions - test_labels)
# # Print out the mean absolute error (mae)
# print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Extract one tree from the random forest
single_tree = rf.estimators_[0]  # Get the first decision tree

# Plot the tree
plt.figure(figsize=(20, 10))  
plot_tree(
    single_tree,
    feature_names=feature_list,  
    class_names=[str(cls) for cls in rf.classes_], 
    filled=True, 
    rounded=True  
)
plt.title("Predicing if Commuting was the #1 Place to Listen to Music ")
plt.show()
```

```python
correct = 0
for idx, val in enumerate(predictions):
    if val == test_labels[idx]:
        correct += 1

```

```python
accuracy = correct / len(predictions)
print(accuracy)
```
